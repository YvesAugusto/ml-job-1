```python
import sklearn
import numpy as np
import pandas as pd
from pathlib import Path
```

# Leitura dos datasets

```python
def parse_data(data):
    Y = data[data.columns[-1]]
    X = data.drop([data.columns[-1]], axis=1)
    
    return X, Y

DATASETS = []

NAMES = [
    "iris_bin_1", "rip"
]
BASE_DIR = Path("../datasets/real/")

for name in NAMES:
    path = BASE_DIR / name / "data.csv"
    d = pd.read_csv(str(path), header=None)
    DATASETS.append(d)

NAMES = [
    "cano", "two_moon", "ying_yang"
]
BASE_DIR = Path("../datasets/artificial/")

for name in NAMES:
    path = BASE_DIR / name / "data.csv"
    d = pd.read_csv(str(path), header=None)
    DATASETS.append(d)
```

# MÃ©tricas

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings("ignore")

def make_grid_params_rbf():
    
    Cs = np.logspace(-1, 2.1, 6)
    Gammas = np.logspace(-2, 2.1, 6)
    kernel = np.array(["rbf"])

    grid = {
        "rbf__C": Cs,
        "rbf__gamma": Gammas,
        "rbf__kernel": kernel
    }
    
    return grid

def make_grid_params_poly():
    
    Cs = np.logspace(-1, 2.1, 4)
    Gammas = np.logspace(-2, 1.6, 4)
    kernel = np.array(["poly"])

    grid = {
        "poly__C": Cs,
        "poly__gamma": Gammas,
        "poly__kernel": kernel,
        "poly__degree": [2, 3, 4]
    }
    
    return grid

def make_grid_params_linear():
    
    Cs = np.logspace(-1, 2.1, 4)
    Gammas = np.logspace(-2, 1.6, 4)
    kernel = np.array(["linear"])

    grid = {
        "linear__C": Cs,
        "linear__gamma": Gammas,
        "linear__kernel": kernel,
    }
    
    return grid

def score(y_true, y_pred):
    
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)
    
    # accuracy
    acc = np.mean(y_true == y_pred)
    
    # preliminar variables
    pos_label = np.unique(y_true)[0]
    neg_label = np.unique(y_true)[1]
    
    # sensitivity
    pos_index = (y_true == pos_label)
    sens = np.mean(y_true[pos_index] == y_pred[pos_index])
    
    # specificity
    neg_index = (y_true == neg_label)
    spec = np.mean(y_true[neg_index] == y_pred[neg_index])
    
    return acc, sens, spec
```

# SVM - RBF

```python
from sklearn.decomposition import PCA

plot_colors = "ryb"
grid = make_grid_params_rbf()
for (i, dataset) in enumerate(DATASETS):
    
    X, Y = parse_data(dataset)
    ACCS = []
    SENS = []
    SPECS = []
        
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
    scaler = StandardScaler()
    pipeline = Pipeline(steps = [("scaler", scaler), ("rbf", SVC())])
    if len(X.columns) == 4:
        pca = PCA(n_components=2)
        pipeline = Pipeline(steps = [("pca", pca), ("scaler", scaler), ("rbf", SVC())])

    gscv = GridSearchCV(
        estimator=pipeline, param_grid=grid, cv=5
    )
    gscv.fit(X_train, y_train)
    best_params = {key.split("__")[-1]: value for key, value in gscv.best_params_.items()}
    
    for _ in range(5):
        X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
        if len(X.columns) == 4:
            pca = PCA(n_components=2)
            X_train = pca.fit_transform(X_train)
            X_test = pca.transform(X_test)

        # ETAPA DE TREINAMENTO
        clf = SVC(**best_params)
        X_train = scaler.fit_transform(X_train)
        clf.fit(X_train, y_train)

        # ETAPA DE TESTES
        X_test = scaler.transform(X_test)
        Y_pred = clf.predict(X_test)
        
        acc, sens, spec = score(y_test, Y_pred)
        
        ACCS.append(acc)
        SENS.append(sens)
        SPECS.append(spec)
        
    mean_acc, std_acc = np.mean(ACCS), np.std(ACCS)
    mean_sens, std_sens = np.mean(SENS), np.std(SENS)
    mean_spec, std_spec = np.mean(SPECS), np.std(SPECS)
    
    print("Acc: {} +/- {}\nSens: {} +/- {}\nSpec: {} +/- {}".format(
        mean_acc, std_acc, mean_sens, std_sens, mean_spec, std_spec
    ))
    # SUPERFÃCIE DE DECISÃO
    
    if X_test.shape[1] > 2:
        continue
    
    ax = plt.subplot(2, 3, i + 1)
    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)
    DecisionBoundaryDisplay.from_estimator(
        clf,
        X_test,
        cmap=plt.cm.RdYlBu,
        response_method="predict",
        ax=ax,
        xlabel=0,
        ylabel=1,
    )
    
    for i, color in zip(np.unique(y_test), plot_colors):
        idx = np.where(y_test.values == i)[0]
        plt.scatter(
            X_test[idx, 0],
            X_test[idx, 1],
            label=i,
            cmap=plt.cm.RdYlBu,
            edgecolor="black",
            s=15,
        )
    
    plt.show()
```
